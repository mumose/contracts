```python
df = pd.read_csv(full_contract_csv_path).iloc[start_row_idx:stop_row_idx].copy().dropna()
df
```

```python
processor = MarkupLMProcessor.from_pretrained("microsoft/markuplm-base-finetuned-websrc")
processor.parse_html = False
```

```python
nodes = df.highlighted_segmented_text.to_list()
xpaths = df.highlighted_xpaths.to_list()
tokens = []
for i in range(len(df)):
  encoding = processor(nodes=nodes[i:i+1], xpaths=xpaths[i:i+1], return_tensors="pt", truncation=False)
  tokens.append([processor.decode(x) for x in encoding['input_ids'][0]][1:-1])
```

```python
df['tokens'] = tokens
df
```

```python
rename_cols = {
    'segment_number_from_idx':'seg_num',
    'highlighted_segmented_text':'h_text',
}
cols_to_keep = [
    'top',
    'left',
    'width',
    'height',
    'tagged_sequence',
    'seg_num',
    'preds',
    'tokens',
    'h_text'
]
```
